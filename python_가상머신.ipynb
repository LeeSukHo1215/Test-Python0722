# 파이썬의 가상머신이란?
파이썬 컴파일러는 명령어를 바이트코드의 명령문으로 번역하고 하드 디스크의 pyc 파일에 저장한다.</br>
그리고 파이썬 인터프리터는 pyc 파일을 인터프리터 방식으로 기계어로 번역한다. 이러한 인터프리터를 파이썬 가상머신이라고 한다.

##Python 가상 머신(PVM)의 역활
바이트 코드 명령어를 머신 코드로 변환하여 컴퓨터가 해당 머신 코드 명령어를 실행하고 최종 출력을 표시할 수 있도록 하는 것이다.<br/>
<br/>
이 변환을 수행하기 위해 가상머신에는 인터프리터가 장착되어 있는데 인터프리터는 바이트 코드를 기계어 코드로 변환하고 그 기계어 코드를 실행을 위해 컴퓨터 프로세서로 보내는 역활을 한다.

##python 가상 머신 구조
Python은 스택기반 가상머신을 사용</br>
즉 스택 데이터 구조가 중심이 된다.



###1.Compiler
소스파일을 어휘분석 구문분석 의미분석을 수행후 바이트코드로 번역한다

###2.Interpreter
바이트코드를 머신코드로 변역하고 메모리 관리자를 호출하고 메모리 객체를 생성하도록 요청한다.

###3.Memory manager
메모리 객체를 생성 및 명령어에 대한 메모리를 참조한다

###4.Stack
어떠한 데이터를 쌓고, 꺼내는 데이터 구조

#### 스택의 유형

1. 호출 스택 

   값 스택과 블록 스택이 각 프레임의 일부인 프레임 형식 


2. 평가 스택(데이터 스텍)

   모든 프레임에 평가 스택 이 있다. 이 스택은 파이썬 함수가 실행되는 곳이며 
   파이썬 코드를 실행하는 것은 대부분 이 스택을 통해 구성된다.


3. 블록 스택

   모든 프레임에는 블록 스택 이 있다. 이것은 Python에서 특정 유형의 제어 구조를 
   추적하는 데 사용된다.

   (프레임에는 함수의 코드개체에 대한 참조와 지역및 전역 변수에 대한것이 있다)
